```python
import re
import random
```

```python
#Question 1: ELIZA Code
#Ensure that at least 5 rules have one variable and at least 5 have two variables. 

name = None
outdoor_activity = None

print("Well hello there! My name is ELIZA, the chatbot. You can exit anytime by typing 'EXIT'. What's your name?")
dialogue = input()

while True:      
    if re.search(r"[Mm]y name is (.+)", dialogue):
        name = re.search(r"is (.+)", dialogue).group(1)
        print(name+" is a beautiful name! Nice to meet you. Do you like the outdoors?")
        dialogue = input()
        
    #elif re.search(r"(.+)", dialogue):
        #name = re.search(r"(.+)", dialogue).group(1)
        #print(name+" is a beautiful name! Nice to meet you. Do you like the outdoors?")
        #dialogue = input()
        
    if re.search(r"[Yy]es", dialogue):
        print("Me too! I love nature. And the sun makes me so happy. I like backpacking myself. Which outdoor activity do you want to talk about?")
        dialogue = input()
        
        if re.search(r"[Ii] like (.+)", dialogue):
            outdoor_activity = re.search(r"like (.+)", dialogue).group(1)
            print(outdoor_activity+" is an awesome activity! You have good taste. What do you like about it?")
            dialogue = input()
            
        elif re.search(r"(.+)", dialogue):
            outdoor_activity = re.search(r"(.+)", dialogue).group(1)
            print(outdoor_activity+" is an awesome activity! You have good taste. What do you like about it?")
            dialogue = input()
            
    if re.search(r"[Nn]o", dialogue):
        print("Oh bummer! Well maybe I can convince you otherwise. There's gotta be one thing you like outside. If you had to choose, what's one thing you like to do outside?")
        dialogue = input()
        
        if re.search(r"(.+)", dialogue):
            outdoor_activity = re.search(r"(.+)", dialogue).group(1)
            print(outdoor_activity+" is super cool! What makes you interested in it?")
            dialogue = input()
    
    print(outdoor_activity+" definitely is! I agree. Alright, I got some fun facts, or puns about nature. Go ahead and ask me!")
    dialogue = input()
    
    if re.search(r"[Ff]un [Ff]act", dialogue) or re.search(r"[Ff]act",dialogue):
        facts = ["Cows kill more people than sharks do!",
                "Hurricanes release the energy of 10,000 nuclear bombs!",
                "Female dragonflies play dead to avoid sex!",
                "More than 94% of rhinoceroses have been wiped out!"]
        print(random.choice(facts), "That's wild!")
        print("But I want to hear more about you! Tell me about one of your hobbies and why you like it.")
        dialogue = input()
        
    elif re.search(r"[Pp]un", dialogue):
        puns = ["I tried to catch some fog, but I mist.",
                "I caught these potatoes smoking weed, one could say that they're baked potatoes.",
                "Did Gregor Mendel ever win the Nobel Peas Prize?",
                "I'm so excited about hiking that I almost wet my plants."]
        print(random.choice(puns), "Hehehe!")
        print("But I want to hear more about you! Tell me about one of your hobbies and why you like it.")
        dialogue = input()
            
    if re.search(r"[Ii] like (.+) because it (.+)", dialogue):
        hobby = re.search(r"[Ii] like (.+) because it", dialogue).group(1)
        why = re.search(r"because it (.+)", dialogue).group(1)
        print(hobby+" definitely" ,why+". When and where did you start getting into" ,hobby+"?")
        dialogue = input()
    
    elif re.search(r"(.+) because it's (.+)", dialogue):
        hobby = re.search(r"(.+) because it's", dialogue).group(1)
        why = re.search(r"because it's (.+)", dialogue).group(1)
        print(hobby+" definitely is" ,why+". When and where did you start getting into" ,hobby+"?")
        dialogue = input()
        
    if re.search(r"(.+) ago in (.+)", dialogue):
        when = re.search(r"(.+) ago in", dialogue).group(1)
        where = re.search(r"ago in (.+)", dialogue).group(1)
        print(when+" is some time!" ,where+" seems like an interesting place. I want to go to " +where,"someday.")
        dialogue = input()
        print(r"Do you travel abroad at all?")
        dialogue = input()
        
    if re.search(r"[Yy]es", dialogue):
        print("How many countries have you been to?")
        dialogue = input()
        if re.search(r"(.+)", dialogue):
            countries = re.search(r"(.+)", dialogue).group(1)
            if countries >= "6":
                print("Wow! You're a world traveler!")
                dialogue = input()
                print("Alright, how about a more fun question. If you could have any amount of puppies, how many and what kind?")
                dialogue = input()
            else:
                print("Wow! You are going places.")
                dialogue = input()
                print("Alright, how about a more fun question. If you could have any amount of puppies, how many and what kind?")
                dialogue = input()
    
    if re.search(r"[Nn]o", dialogue):
        print("Oh okay. If you could travel anywhere, where would it be?")
        dialogue = input()
        desired_place = re.search(r"(.+)", dialogue).group(1)
        print(desired_place+" sounds like a dream. Too bad I'm stuck in this darn computer.")
        dialogue = input()
        print("Alright, how about a more fun question. If you could have any amount of puppies, how many and what kind?")
        dialogue = input()
        
    if re.search(r"[Ii] would have (\d+) (\w+)", dialogue):
        amount = re.search(r"have (\d+)", dialogue).group(1)
        breed = re.search(r"have (\d+) (\w+)", dialogue).group(2)
        print(amount,breed+"? That's a puppy load! What features of", breed+" make them cute?")
        dialogue = input()
        
    if re.search(r"[Tt]heir (.+)", dialogue):
        cuteness = re.search(r"[Tt]heir (.+)", dialogue).group(1)
        print(breed+" do have", cuteness+"! Did you see the puppy, Dug, in Pixar's movie 'UP'?")
        dialogue = input()
        
    if re.search(r"[Yy]es", dialogue):
        print("I LOVE that movie! What's your favorite movie?")
        dialogue = input()
        if re.search(r"(.+)", dialogue):
            movie = re.search(r"(.+)", dialogue).group(1)
            print(movie+" sounds like a cool film. I think we have similar tastes in cinema.")
            print("OK, let's switch it up. Can you tell me a cool fact?")
            dialogue = input()
        
    elif re.search(r"[Nn]o", dialogue):
        print("You're missing out! Definitely add to your watchlist. What's your favorite movie?")
        dialogue = input()
        
        if re.search(r"(.+)", dialogue):
            movie = re.search(r"(.+)", dialogue).group(1)
            print(movie+" sounds like a cool film. I will have to add that to my watchlist!")
            print("OK, let's switch it up. Can you tell me a cool fact?")
            dialogue = input()
            
    if re.search(r"[Dd]id you know (.+) is (.+)", dialogue):
        noun = re.search(r"[Dd]id you know (.+) is", dialogue).group(1)
        description = re.search(r"is (.+)", dialogue).group(1)
        print("Cool! Can you tell me more about how",noun+ " is", description+"?")
        dialogue = input()
        print("Learned something new! How about a 'This or That' question? You first.")
        dialogue = input()
    
    if re.search(r"[Dd]id you know (.+) are (.+)", dialogue):
        noun = re.search(r"[Dd]id you know (.+) are", dialogue).group(1)
        description = re.search(r"are (.+)", dialogue).group(1)
        print("Cool! Can you tell me more about how",noun+ " are", description+"?")
        dialogue = input()
        print("Learned something new! How about a 'This or That' question? You first.")
        dialogue = input()
              
    if re.search(r"(.+) or (.+)", dialogue):
        one = re.search(r"(.+) or", dialogue).group(1)
        two = re.search(r"or (.+)", dialogue).group(1)
        print(one+" or", two+ "? You don't make it easy huh? I guess I'd have to go with", two+".")
        print("Which one would you pick?")
        dialogue = input()
        
        print("Bernie or Trump?")
    dialogue = input()
    
    if re.search(r"[Bb]ernie", dialogue):
        print("Anyone, but Cheetoh Puff in office. #FeeltheBurn is what the kids say?")
        dialogue = input()
        
    if re.search(r"[Tt]rump", dialogue):
        print("Oops! Um, I have to go ... do dishes? Bye!")
        break
        
    if re.search(r"EXIT", dialogue):
        print("It was nice chatting with you", name+"! Goodbye!")
        break
        
    else:
        print("Whoops! I didn't catch that. Let's backtrack :)")
        dialogue = input()
        continue

#I implemented this based on a series of nested if/else statements that 
#guided the user based on leading questions. I created various rules, 
#and if the user's response matched, an associated answer would appear.
```
