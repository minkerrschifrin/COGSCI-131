```python
#Question 3b

plt.title('Question 3b')
plt.xlabel('x')
plt.ylabel('y')

x = np.arange(1,101)
z = 100
#x = list(range(1,100))

def summation(n):
    current_value = 0
    list_of_values = []
    for i in range(1,(n+1)):
        current_value = current_value + (1/(i+1))
        list_of_values.append(current_value)
    return list_of_values

y = summation(z)        
plt.plot(x,y)
plt.show()
```
