```python
#Question 2

plt.title('Question 2')
plt.xlabel('Trial Number')
plt.ylabel('Association')

alpha, beta = 0.2, 0.1
association_bell_and_food = np.array([])
light = 0.8
bell = 0.0

i = 1
while (i <= 20):
    bell = bell + alpha * beta * (1 - light - bell)
    association_bell_and_food = np.append(association_bell_and_food, bell)
    i += 1
plt.plot(association_bell_and_food)
```

```python
#Question 3a

plt.title('Question 3a')
plt.xlabel('Trial Number')
plt.ylabel('Association')

alpha, beta = 0.2, 0.1
alternate_trials = ['food', 'no_food']*500

association = np.array([])

Vx = 0.0
   
for trial in alternate_trials:
    if trial == 'food':
        Vx = Vx + alpha * beta * (1 - Vx)
        association = np.append(association, Vx)
    elif trial == 'no_food':
        Vx = Vx - alpha * beta * (Vx)
        association = np.append(association, Vx)

x = np.arange(len(association))            
plt.plot(x, association)
print("Explanation:")
print("Considering that we are alternating trials between food and no food, \nit makes sense that there are more trials than without alternation. \nAdditionally, the association tops off at 0.5, as opposed to 1, which \nbell and food consistently would provide. Alternating food and no food \ntrials, lowers the association.")
```
