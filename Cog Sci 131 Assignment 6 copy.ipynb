```python
import random
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
#Question 1

trials = 10000

positive = []
negative = []

positive = np.random.uniform(0, 10, trials)
negative = np.random.uniform(-10, 0, trials)

def contains(x):
    heights = []
    
    for i in range(0, trials):
        length = positive[i] - negative[i]
        height = 1/length
        heights.append(height)

    height = 0
    probability = 0
    
    for i in range(0, trials):
        if x < positive[i] and x >= negative[i]:
            height = height + heights[i] 
            dividend = height
            denominator = np.sum(heights)
            probability = dividend/denominator
    return probability
```

```python
#Question 2

print(contains(1))
```

```python
#Question 3

plt.title('Probability of X-Values: 0 to 10')
plt.xlabel('X-Values')
plt.ylabel('Probability')

def ten():
    probability = [] 
    steps = []
        
    for i in range(0, 11):
        probability.append(contains(i))
        steps.append(i)
    plt.plot(steps, probability)
    plt.show()
    
ten()
```

The function has a negative slope as it moves towards 10. Considering that the probability, when x = 0, is 1, it makes sense that 0 has the highest probability; 0 is guaranteed to be an x-value. But as x approaches 10, the likelihood of an x-value being in that region decreases.

Plotting a logarithmic y axis (and looking for a straight line) allows one to check whether a curve has an exponential decrease because logarithm and exponentials are inverses of one another. A straight decreasing line shows exponential decay and a straight increasing line shows exponential growth.

```python
#Question 5

plt.title('Probability of X-Values: -5 to 5')
plt.xlabel('X-Values')
plt.ylabel('Logarithmic Probability')
plt.yscale('log')

def five():
    probability = [] 
    steps = []
        
    for i in range(-5, 6):
        probability.append(contains(i))
        steps.append(i)
    plt.plot(steps, probability)
    plt.show()
    
five()
```

```python
plt.title('Probability of X-Values: -10 to 10')
plt.xlabel('X-Values')
plt.ylabel('Logarithmic Probability')
plt.yscale('log')

def log_ten():
    probability = [] 
    steps = []
        
    for i in range(-10, 11):
        probability.append(contains(i))
        steps.append(i)
    plt.plot(steps, probability)
    plt.show()
    
log_ten()
```

Similar to the reasoning for #3, the probabilty is 1 when x = 0, and when the x-values stray (positively or negatively) from 0, the likelihood of the x-value being in those regions decreases. This also shows conditional probability distribution which enforces that there is a certain pattern for the probability distribution of y, considering that there are given x-values.
